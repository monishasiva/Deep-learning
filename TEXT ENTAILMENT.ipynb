# a. BASIC TEXT ENTAILMENT USING SIMPLE RULE-BASED METHODS

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


# Sample premise and hypothesis
premise = "The cat is sitting on the mat."
hypothesis = "A cat is on the mat."

# Set of English stopwords
stop_words = set(stopwords.words('english'))

# Preprocess function: tokenize, lowercase, remove stopwords and non-alpha tokens
def preprocess(text):
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalpha() and word not in stop_words]

# Tokenize and clean premise and hypothesis
premise_tokens = preprocess(premise)
hypothesis_tokens = preprocess(hypothesis)

# Rule-based entailment check:
# If all hypothesis tokens are present in the premise tokens -> entailment
common_words = set(hypothesis_tokens).intersection(set(premise_tokens))

if len(common_words) == len(set(hypothesis_tokens)):
    print("Entailment")
else:
    print("Not Entailment")
```

# b. NATURAL LANGUAGE INFERENCE WITH BERT

```python
from transformers import pipeline

# Load NLI pipeline
nli_model = pipeline("zero-shot-classification", model="roberta-large-mnli")

# Sample premise and hypothesis
premise = "A man is playing a guitar."
hypothesis = "A person is playing an instrument."

# In zero-shot-classification, you provide a candidate label — here, hypothesis acts as the label against premise
result = nli_model(premise, candidate_labels=[hypothesis])

print(result)
```

# c. SENTENCE PAIR CLASSIFICATION USING SIAMESE NETWORKS

```python
from sentence_transformers import SentenceTransformer, util

# Load pretrained model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Sample sentences
sentence1 = "A man is eating food."
sentence2 = "A person is consuming a meal."

# Encode sentences
embedding1 = model.encode(sentence1, convert_to_tensor=True)
embedding2 = model.encode(sentence2, convert_to_tensor=True)

# Compute cosine similarity
similarity = util.pytorch_cos_sim(embedding1, embedding2)

print(f"Cosine Similarity: {similarity.item()}")

# Simple threshold to determine entailment (example threshold = 0.75)
if similarity.item() > 0.75:
    print("Entailment")
else:
    print("Not Entailment")
```
